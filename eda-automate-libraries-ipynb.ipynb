```python
#sixth way using lida
# !pip install lida

# 📊 Pros and Cons of LIDA (Language Interface for Data Analysis)

# ✅ Pros:
# - Natural language interface: You can ask questions like "What's the average age?" instead of writing code.
# - Fast EDA: Quickly generate charts and summaries from a dataset.
# - Reduces need for coding: Great for beginners or non-programmers.
# - Visualization suggestions: Recommends useful charts based on your data.
# - Interactive exploration: Supports human-AI collaboration for analysis.

# ❌ Cons:
# - Limited to common questions: May not handle complex or custom analyses well.
# - Requires clean data: Struggles with messy or highly unstructured datasets.
# - Dependent on model accuracy: May give wrong answers or visualizations if misunderstood.
# - Lacks full flexibility: Can't replace pandas/Matplotlib/seaborn for advanced users.
# - Not fully supported in all environments (like Kaggle or restricted notebooks).

# 👉 Best used for: Quick EDA, idea generation, teaching, and automating basic analysis tasks.
```
